```python
import pandas as pd
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
import matplotlib.dates as mdates
import seaborn as sns

matplotlib.font_manager._rebuild()
import numpy as np
from matplotlib.dates import DateFormatter
matplotlib.rcParams['figure.figsize'] = [16,16]
```

```python
df = pd.read_csv('udzialy.csv', delimiter = ";")
for col in df.columns[2:]:
    df[col] = df[col].str.rstrip('%').astype('float') / 100.0
df.columns=['od', 'do', 'D-P (H-W)', 'P-D (W-H)', 'D-N (H-E)', 'N-D (E-H)', 'D-I (H-O)', 'I-D (O-H)', 'NZD (NHB)']
```

```python
fig, axes = plt.subplots(nrows=4, ncols=2, sharex='all', sharey='all')
motywacje = df.columns[2:]
i=0
for motywacja in motywacje:
    ax = axes.flatten()[i]
    df[motywacja].plot(color='black',title=motywacja, ax = ax)
    if i == 4:
        ax.set_ylabel('udzial podrozy (share)')
    if i>3:
        ax.set_xlabel('godzina (hour)')
      
    i+=1

for motywacja in motywacje:    
    df[motywacja].plot(title='wszystkie (all)', ax = axes.flatten()[i])
    plt.xlabel('godzina (hour)')
fig.suptitle('Rozklad dobowy podrozy (temporal trip ditribution)')


```

```python
matplotlib.rcParams['figure.figsize'] = [16,4]
df[motywacje].plot()
plt.ylabel('udzial podrozy (share)')
plt.xlabel('godzina (hour)')
plt.savefig('all.png', dpi = 320)
plt.show()
```
